```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("/content/diabetes.csv")
print(len(data))
print(data.head())
```

```python
zeros_nota_accept=["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
for column in zeros_nota_accept:
  data[column]=data[column].replace(0,np.NaN)
  mean=int(data[column].mean(skipna=True))
  data[column]=data[column].replace(np.NaN,mean)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
X=data.drop("Outcome",axis=1)
y=data["Outcome"]
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
Scaler=StandardScaler()
X_train=Scaler.fit_transform(X_train)
X_test=Scaler.fit_transform(X_test)
```

```python
model=KNeighborsClassifier(n_neighbors=11,p=2,metric="euclidean")
model.fit(X_train,y_train)
```

```python
pridiction=model.predict(X_test)
print(pridiction)
```

```python
model_score=model.score(X_test,y_test)
print(model_score)
```

```python

cm=confusion_matrix(y_test,pridiction)
print("Confusion Matrix",cm)
print("F1 Score",f1_score(y_test,pridiction))
print("Accuracy Score",accuracy_score(y_test,pridiction))
```
